```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Function to split train test set:

```python
def split_dataset(X, y, ratio):
    arr = np.arange(X.size)
    np.random.shuffle(arr)
    num_train = int(ratio * X.size)
    X_train = X[arr[:num_train]]
    y_train = y[arr[:num_train]]
    X_test = X[arr[num_train:]]
    y_test = y[arr[num_train:]]
    return X_train, X_test, y_train, y_test
```

### Creating data and setting hyper parameters for regularized polynomial regression

```python
learning_rate = 0.001
num_epochs = 1000
reg_lambda = 0.

X_data = np.linspace(-1,1,101)

num_coeffs = 9
y_data_params = [0.]*num_coeffs
y_data_params[2] = 1.
y_data = 0
for i in range(num_coeffs):
    y_data += y_data_params[i]*np.power(X_data, i)
y_data += np.random.randn(*X_data.shape)*0.3
X_train, X_test, y_train, y_test = split_dataset(X_data, y_data, 0.3)
```

### Setting up tensorflow model with regularized cost function:

```python
X = tf.placeholder("float")
y = tf.placeholder("float")

def model(X, w):
    terms = []
    for i in range(num_coeffs):
        terms.append(tf.multiply(w[i], tf.pow(X, i)))
    return tf.add_n(terms)

w = tf.Variable([0.]*num_coeffs, name = "parameters")
y_model = model(X, w)
cost = tf.divide(
                tf.add(tf.reduce_sum(tf.square(y - y_model)), tf.multiply(reg_lambda, tf.reduce_sum(tf.pow(w, 2)))),
                2*X_train.size)

train_op = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost)

sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)

for reg_lambda in np.linspace(0,1,100):
    for epoch in range(num_epochs):
        sess.run(train_op, feed_dict = {X:X_train, y:y_train})
    final_cost = sess.run(cost, feed_dict = {X:X_test, y: y_test})
    print "reg_lambda : ", reg_lambda
    print "final cost : ", final_cost
    
sess.close()
```
